```python
x = df["no2"]      # change only if column name differs
x = pd.to_numeric(x, errors='coerce')
x = x.dropna()
```

```python
x = x[x >= 0]
x = x[x <= 500]
x = x.to_numpy()
len(x)
```

```python
a = 0.10
b = 0.9

z = x + a * np.sin(b * x)
```

```python
mu = np.mean(z)
mu
```

```python
sigma2 = np.var(z)
sigma2
```

```python
sigma = np.sqrt(sigma2)

lam = 1/(2*sigma2)
c = 1/(sigma*np.sqrt(2*np.pi))

lam, mu, c
```
